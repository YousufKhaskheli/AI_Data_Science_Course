```python
def get_formatted_name(first_name, last_name, middle_name=''):
  if middle_name:
    full_name = first_name + ' ' + middle_name + ' ' + last_name
  else:
    full_name = first_name + ' ' + last_name
  return full_name.title()
```

#Simple Bank


```python
import random
all_accounts=[]
```

```python
all_accounts
```

```python
def open_account(cnic,initial_deposit=0):
  account={'cnic':cnic, 'balance':initial_deposit}
  account_title = input("Enter title of account")
  account_number = random.randint(10000,99999)
  account_pin = random.randint(1000,9999)
  account['title'] = account_title
  account['pin'] = account_pin
  account['account_number'] = account_number
  all_accounts.append(account)
  return f""" Thank you for opening an account with us!
            {account}"""
```

```python
print(open_account(423012468999, 10000))
```

```python
def check_balance(account_number, pin):
  for account in all_accounts:
    if account['account_number'] == account_number:
      if account['pin'] == pin:
        return f"Your account balance is Rs:{account['balance']}"
      else:
        return "Invalid pin"
  else:
    return "Invalid account number"
```

```python
check_balance(98405,4973)
```

```python
check_balance(98405,49)
```

```python
def cash_withdrawal(account_number, pin, amount):
  for account in all_accounts:
      if account['account_number'] == account_number:
        if account['pin'] == pin:
          if account['balance']>=amount:
            account['balance']-=amount
            return f"You have withdrawn Rs:{amount} succesfully"
          else:
            return "Induficient amount"
        else:
          return "Invalid pin"
  else:
    return "Invalid account number"
```

```python
def cash_deposit(account_number, amount):
  for account in all_accounts:
      if account['account_number'] == account_number:
          if account['balance']>=amount:
            account['balance']+=amount
            return f"You have deposited Rs:{amount} succesfully"
          else:
            return "Indsufficient amount"
  else:
    return "Invalid account number"
```

```python
cash_withdrawal(98405,4973,2500)
```

```python
check_balance(98405,4973)
```
