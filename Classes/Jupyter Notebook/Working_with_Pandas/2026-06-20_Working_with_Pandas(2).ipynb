```python
def calc_tax(salary):
  if salary >= 200000:
    return salary*.20
  elif salary >= 100000:
    return salary*.10
  elif salary >= 75000:
    return salary*.075
  elif salary >= 50000:
    return salary*.05
  else:
    return salary*.010
```

```python
# using function
df1["Tax"] = df1.Salary.apply(calc_tax)
df1
```

```python
#using lambda
df1["Tax"] = df1.Salary.apply(lambda sal:
                              sal*0.20 if sal>=200000
                              else sal*.15 if sal>=150000
                              else sal*.10 if sal>= 100000
                              else sal*.075 if sal>=75000
                              else sal*.05 if sal>=50000
                              else sal*.010)
df1
```

##Querry Filters

```python
# need the data of all managers
df1[df1.Designation=='Manager']
```

```python
# need all the employees of IT
df1[df1.Department=='IT']
```

```python
# all employees ages more than 30
df1[df1.Age>30]
```

```python
# Number of employees in every department
df1.Department.value_counts()
```

```python
df1.Department.replace('Account','Accounts',inplace=True)
```

```python
df1.Department.value_counts()
```

```python
# what is the total salary of Managers
df1[df1.Designation=='Manager']['Salary'].sum()
```

```python
# what is the average salary of IT employees
df1[df1.Department=='IT']['Salary'].mean()
```

```python
# what is the maximum salary in accounts department
df1[df1.Department=='Accounts']['Salary'].max()
```

```python
# what is the min age of Manager
df1[df1.Designation=='Manager']['Age'].min()
```

```python
# what is the average age of employee in HR
df1[df1.Department=='HR']['Age'].mean()
```

```python
# All officers in accounts
df1[(df1.Designation=='Officer') & (df1.Department=='Accounts')]
```

```python
# All Managers having salary greater than or equal to 100K
df1[(df1.Designation=='Manager') & (df1.Salary>=10000)]
```

```python
# all the employees with salalry greater than average salary
df1[df1.Salary>df1.Salary.mean()]
```
