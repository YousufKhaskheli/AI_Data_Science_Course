#Inheritance


```python
class Car:
    def __init__(self,make,model,year,color,seating, engine):# parameter
        # object variables/attributes
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.seating = seating
        self.engine = engine

   # methods
    def car_details(self):
        return f"""
                    {self.make}
                    {self.model}
                    {self.year}
                    {self.color}
                    {self.seating}
                    {self.engine}"""
    def start(self):
        return f"{self.model}{self.make} is starting!!!!"
    def stop(self):
        return  f"{self.model}{self.make} is stopping!!!!"
    def fill_gas_tank(self):
        return f"{self.model}{self.make} is filling gass tank!!!!"
    def run(self):
        return f"{self.model}{self.make} is running fast!!!!"
```

```python
class elecricCar(Car):
  battery = '40Amp'
```

```python
class ElectricCar(Car):
  def __init__(self,make,model,year,color,seating,engine,battery,owner,variant,mode,price):
    self.battery = battery
    self.owner = owner
    self.variant = variant
    self.mode = mode
    self.price = price
    super().__init__(make,model,year,color,seating,engine)
```

```python
ec1 = ElectricCar('Tesla','Roadster',2025,'White','Leather','V8','120,000mWh','Yousuf','V2','Sports','250,000USD')
```

```python
ec1.car_details()
```

```python
class Restaurant():
  def __init__(self,name,cuisine,num_served=0):
    self.name = name
    self.cuisine = cuisine
    self.num_served = num_served
  def describe_restaurant(self):
    return f"Restaurant name: {self.name}, Cuisine type: {self.cuisine}"
  def open_restaurant(self):
    return f"The Restaurant {self.name} is open"
  def set_num_served(self,new_num_served):
    self.num_served = new_num_served
  def increment_num_served(self,amount=1):
    self.num_served += amount
```
