### Slicing Lists

Slicing allows you to create a new list containing a subset of elements from an existing list. It uses the following syntax:

`list[start:end:step]`

*   **`start`**: The index where the slice begins (inclusive). If omitted, it defaults to the beginning of the list (0).
*   **`end`**: The index where the slice ends (exclusive). If omitted, it defaults to the end of the list.
*   **`step`**: The increment between elements in the slice. If omitted, it defaults to 1.

```python
students = ['Rehman', 'Yousuf', 'Umer', 'Ali', 'Ilyas', 'Abid']
print(f"Original list: {students}")

# Example 1: Basic slice (elements from index 1 up to, but not including, index 4)
print(f"students[1:4]: {students[1:4]}")

# Example 2: Slice from the beginning to a specific index
print(f"students[:3]: {students[:3]}")

# Example 3: Slice from a specific index to the end
print(f"students[2:]: {students[2:]}")

# Example 4: Copying the entire list
print(f"students[:]: {students[:]}")

# Example 5: Slicing with a step (every second element)
print(f"students[::2]: {students[::2]}")

# Example 6: Reverse the list using slicing
print(f"students[::-1]: {students[::-1]}")

# Example 7: Negative indices (from the end of the list)
print(f"students[-3:-1]: {students[-3:-1]}")
```

```python
students = ['Rehman','Yousuf','Umer','Ali','Rehman','Abid']
print(f"Original students list: {students}")

# 1. append(item): Adds an item to the end of the list.
students.append('Imran')
print(f"After append('Imran'): {students}")

# 2. insert(index, item): Inserts an item at a specified index.
students.insert(2, 'Sara')
print(f"After insert(2, 'Sara'): {students}")

# 3. extend(iterable): Appends elements from an iterable (e.g., another list) to the end of the current list.
new_students = ['Hassan', 'Fatima']
students.extend(new_students)
print(f"After extend(['Hassan', 'Fatima']): {students}")

# 4. remove(item): Removes the first occurrence of a specified item.
students.remove('Rehman') # Removes the first 'Rehman'
print(f"After remove('Rehman'): {students}")

# 5. pop(index=-1): Removes and returns the item at a given index. If no index, removes last item.
popped_student = students.pop()
print(f"After pop() (removed {popped_student}): {students}")
popped_at_index = students.pop(0) # Removes the first item
print(f"After pop(0) (removed {popped_at_index}): {students}")

# 6. index(item, start=0, end=len(list)): Returns the index of the first occurrence of an item.
try:
    index_of_ali = students.index('Ali')
    print(f"Index of 'Ali': {index_of_ali}")
except ValueError:
    print("'Ali' not found in the list")

# 7. count(item): Returns the number of times an item appears in the list.
rehman_count = students.count('Rehman')
print(f"Count of 'Rehman': {rehman_count}")

# 8. sort(key=None, reverse=False): Sorts the list in ascending order by default.
students_to_sort = ['Zoe', 'Alice', 'Bob', 'Charlie']
students_to_sort.sort()
print(f"Sorted list: {students_to_sort}")
students_to_sort.sort(reverse=True)
print(f"Reverse sorted list: {students_to_sort}")

# 9. reverse(): Reverses the order of elements in the list.
students.reverse()
print(f"After reverse(): {students}")

# 10. copy(): Returns a shallow copy of the list.
students_copy = students.copy()
print(f"Copied list: {students_copy}")

# 11. clear(): Removes all items from the list.
students.clear()
print(f"After clear(): {students}")
```

### Deleting Members from a List:

*   **`del list[index]` or `del list[start:end]`**: Deletes items permanently by their index or a slice. It does not return the deleted item(s).
    *Example: `del my_list[0]` or `del my_list[1:3]`*
*   **`pop(index=-1)`**: Removes and returns the item at a given index. If no index is specified, it removes and returns the last item.
    *Example: `removed_item = my_list.pop()` or `removed_item = my_list.pop(2)`*
*   **`remove(value)`**: Removes the *first* occurrence of a specified value from the list. It raises a `ValueError` if the value is not found.
    *Example: `my_list.remove('old_element')`*
*   **`clear()`**: Removes all items from the list, making it empty.
    *Example: `my_list.clear()`*

#String: **Immutable**

### Common String Methods:

*   **`len(string)`**: Returns the length (number of characters) of the string.
    *Example: `len("hello")` would return `5`*
*   **`string.upper()`**: Returns a new string with all characters converted to uppercase.
    *Example: `"Hello".upper()` would return `"HELLO"`*
*   **`string.lower()`**: Returns a new string with all characters converted to lowercase.
    *Example: `"Hello".lower()` would return `"hello"`*
*   **`string.capitalize()`**: Returns a new string with its first character capitalized and the rest lowercased.
    *Example: `"hello world".capitalize()` would return `"Hello world"`*
*   **`string.title()`**: Returns a new string where the first letter of each word is capitalized, and the rest are lowercased.
    *Example: `"hello world".title()` would return `"Hello World"`*
*   **`string.strip()`**: Returns a new string with leading and trailing whitespace removed.
    *Example: `"  hello  ".strip()` would return `"hello"`*
*   **`string.replace(old, new)`**: Returns a new string with all occurrences of `old` substring replaced by `new` substring.
    *Example: `"hello world".replace("world", "python")` would return `"hello python"`*
*   **`string.split(separator)`**: Returns a list of substrings by splitting the string at each occurrence of the `separator`.
    *Example: `"apple,banana,cherry".split(",")` would return `['apple', 'banana', 'cherry']`*
*   **`string.join(iterable)`**: Concatenates the elements of an iterable (e.g., a list of strings) using the string as a separator.
    *Example: `"-".join(["1", "2", "3"])` would return `"1-2-3"`*
*   **`string.find(substring)`**: Returns the lowest index in the string where the `substring` is found. Returns `-1` if not found.
    *Example: `"hello".find("ll")` would return `2`*
*   **`string.startswith(prefix)`**: Returns `True` if the string starts with the specified `prefix`, otherwise `False`.
    *Example: `"hello".startswith("he")` would return `True`*
*   **`string.endswith(suffix)`**: Returns `True` if the string ends with the specified `suffix`, otherwise `False`.
    *Example: `"hello".endswith("lo")` would return `True`*
*   **`string.count(substring)`**: Returns the number of non-overlapping occurrences of `substring` in the string.
    *Example: `"banana".count("na")` would return `2`*
*   **`string.isalpha()`**: Returns `True` if all characters in the string are alphabetic and there is at least one character, `False` otherwise.
*   **`string.isdigit()`**: Returns `True` if all characters in the string are digits and there is at least one character, `False` otherwise.
*   **`string.isalnum()`**: Returns `True` if all characters in the string are alphanumeric (letters or numbers) and there is at least one character, `False` otherwise.
