```python
#  tag numbers as even or odd
numbers= [4, 1, 25, 36, 49, 64, 81, 121, 0, 16, 100]
for n in numbers:
    if n%2==0:
        print(n, 'even')
    else:
        print(n,'odd')
```

```python
# 1 units of electric is Rs: 50.
#  number of units consumed , total bill but also add 10tax.
price_per_unit = 50
units = int(input("Enter number of units consumed: "))
total = units * price_per_unit
tax = total * .10

bill = total + tax
print(f"total {total}")
print(f"tax {tax}")
print(f"Your total bill amount is {bill}")
```
