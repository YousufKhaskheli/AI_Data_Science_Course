#Acess Modifier
- Public

- Protected


## Types Of Access Modifiers (Simplified for Python)

Think of access modifiers like rules for who can see or change a variable or function within your code.

### Public (No special prefix)

*   **What it means:** Anyone can access it, both from inside and outside the class. It's like an item you leave on your desk for everyone to use.
*   **When to use:** For things you want other parts of your code to interact with directly.
*   **Python example:** `self.name`, `self.marks`. You can directly do `s1.name = 'New Name'` or `print(s1.marks)`.

### Protected (Single underscore: `_variable_name`)

*   **What it means:** It's a hint to other programmers: "Please don't touch this directly from outside the class or its children classes." It's like a file in your personal folder – you can access it, but it's generally understood you shouldn't mess with it unless you know what you're doing or you're a child of this class.
*   **When to use:** For internal details of your class or for methods/attributes that child classes might need to use or override.
*   **Python example:** `self._internal_data`. Python won't stop you from accessing `s1._internal_data`, but it's a convention to avoid it.

### Private (Double underscore: `__variable_name`)

*   **What it means:** Python makes it a bit harder to access this from outside the class. It's like a locked drawer in your desk – you can't just reach in. Python 'mangles' its name, so you'd need to know the hidden name to get to it.
*   **When to use:** For things that are truly internal to your class and should almost never be changed or seen from the outside. This helps prevent accidental changes.
*   **Python example:** `self.__secret_attribute`. If you define `__secret_attribute`, you can't just say `s1.__secret_attribute`. You'd technically need to do `s1._student__secret_attribute`, which shows it's meant to be private.

```python
class student():
  def __init__(self,name,roll,course,email,marks):
# public instance variable
    self.name=name
    self.roll=roll
    self.course=course
    self.email=email
#   protected instance variable
    self._marks=marks
  def show_proile(self):
    print("Student profile")
    print(f"Name: {self.name}")
    print(f"Roll: {self.roll}")
    print(f"Course: {self.course}")
    print(f"Email: {self.email}")
    print(f"Marks: {self._marks}")
  def fetch_marks(self):
    return self._marks
  def set_marks(self,new_marks):
    if (new_marks >= 0) and (new_marks <= 100):
      self._marks = new_marks
    else:
      return "Invalid marks"


```

```python
class patient:
    def __init__(self,name,age,id,disease,bld_grp):
        self.name=name
        self.age=age
        self.id=id
        self.disease=disease
        self.bld_grp=bld_grp
    def patient_profile(self):
      print(f"Patient {self.id}")
      print(f"Name: {self.name}")
      print(f"Age: {self.age}")
      print(f"Disease: {self.disease}")
      print(f"Blood Group: {self.bld_grp}")
```
