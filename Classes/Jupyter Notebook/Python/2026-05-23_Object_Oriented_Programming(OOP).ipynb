```python
class Car:    
    def __init__(self,make,model,year,color,seating, engine):# parameter
        # object variables/attributes
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.seating = seating
        self.engine = engine
        
   # methods
    def car_details(self):
        return f""" 
                    {self.make}
                    {self.model}
                    {self.year}
                    {self.color}
                    {self.seating}
                    {self.engine}
                    """
    def start(self):
        return f"{self.model}{self.make} is starting!!!!"
    def stop(self):
        return  f"{self.model}{self.make} is stopping!!!!"
    def fill_gas_tank(self):
        return f"{self.model}{self.make} is filling gass tank!!!!"
    def run(self):
        return f"{self.model}{self.make} is running fast!!!!"  
```

```python
c1 = Car('Honda','City',2025,'white',5,'1300CC')# argument
```

```python
print(c1.car_details())
```

```python
print(c1.start())
```

```python
print(c1.run())
```

```python
print(c1.stop())
```
