# While Loop

```python
amount=10000
while amount>0:
  print(amount, "Do Shopping")
  amount-=1000
```

```python
count=0
while count<=10:
  print(count, "count")
  count+=1
```

```python
num = int(input("Enter a number (n):"))
sum_n = 0
counter = 1
while counter <= num:
    sum_n += counter
    counter += 1
print(f"The sum of the first {num} natural numbers is: {sum_n}")
```

```python
# run a while loop 10 times every time while doubles the number
num=1
while num<=10:
  print(f"Double of number {num} is: {num*2}")
  num+=1
```

```python
# run a while loop 10 times every time if number is even double it and if its odd triple it
num=0
while num<=10:
  if num%2==0:
      print(f"Double of number {num} is: {num*2}")
  else:
      print(f"Double of number {num} is: {num*3}")
  num+=1

```

```python
def calculator(num1,num2,operation):
  if operation=="+":
    print(f"Addition of {num1} and {num2} is: {num1+num2}")
  elif operation=="-":
    print(f"Subtraction of {num1} and {num2} is: {num1-num}")
  elif operation=="*":
    print(f"Multiplication of {num1} and {num2} is: {num1*num2}")
  elif operation=="/":
    print(f"Division of {num1} and {num2} is: {num1/num2}")
  else:
    print("Invalid operation")
```

```python
calculator(12,34,'/')
```

```python
def name_analyzer(name):
  vowels = ['a','e','i','o','u','A','E','I','O','U']
  vowels_count = 0
  upper_count = 0
  lower_count = 0
  for char in name:
    if char in vowels:
      vowels_count+=1
    if char.isupper():
      upper_count+=1
    if char.islower():
      lower_count+=1
  print(f"""
         Vowels: {vowels_count}
         upper : {upper_count}
         lower : {lower_count}""")
```

```python
name_analyzer("Yousuf Khaskheli")
```
